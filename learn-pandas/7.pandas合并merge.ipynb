```python
import pandas as pd 
```

```python
left  = pd.DataFrame({
                        'key':['k0', 'k1','k2', 'k3'],
                        'A':['A0', 'A1', 'A2', 'A3'],
                        'B':['B0', 'B1', 'B2', 'B3']
                     })
right  = pd.DataFrame({
                        'key':['k0', 'k1','k2', 'k3'],
                        'C':['C0', 'C1', 'C2', 'C3'],
                        'D':['D0', 'D1', 'D2', 'D3']
                     })
print(left)
print(right)
```

```python
res = pd.merge(left, right, on='key')
res
```

```python
left  = pd.DataFrame({
                        'key1':['k0', 'k0','k1', 'k2'],
                        'key2':['k0', 'k1','k0', 'k1'],
                        'A':['A0', 'A1', 'A2', 'A3'],
                        'B':['B0', 'B1', 'B2', 'B3']
                     })
right  = pd.DataFrame({
                        'key1':['k0', 'k1','k1', 'k3'],
                        'key2':['k0', 'k0','k0', 'k0'],
                        'C':['C0', 'C1', 'C2', 'C3'],
                        'D':['D0', 'D1', 'D2', 'D3']
                     })
print(left)
print(right)
```

```python
# how 默认是inner,  取值  left, right, inner, outer
# outer 相当于取并集， inner 相当于取交集
res = pd.merge(left, right, on = ['key1', 'key2'], how='outer') 
res
```

```python
# how 默认是inner,  取值  left, right, inner, outer
# outer 相当于取并集， inner 相当于取交集
res = pd.merge(left, right, on = ['key1', 'key2'], how='inner') 
res
```

```python
# how 默认是inner,  取值  left, right, inner, outer
# outer 相当于取并集， inner 相当于取交集
res = pd.merge(left, right, on = ['key1', 'key2'], how='left') 
res
```

```python
# how 默认是inner,  取值  left, right, inner, outer
# outer 相当于取并集， inner 相当于取交集
res = pd.merge(left, right, on = ['key1', 'key2'], how='right') 
res
```

```python
# how 默认是inner,  取值  left, right, inner, outer
# outer 相当于取并集， inner 相当于取交集
# indicator 显示合并的信息
res = pd.merge(left, right, on = ['key1', 'key2'], how='outer', indicator=True) 
res
```

```python
# how 默认是inner,  取值  left, right, inner, outer
# outer 相当于取并集， inner 相当于取交集
# indicator 显示合并的信息
res = pd.merge(left, right, on = ['key1', 'key2'], how='outer', indicator='indicator_info') 
res
```

```python
left  = pd.DataFrame({
                        'A':['A0', 'A1', 'A2'],
                        'B':['B0', 'B1', 'B2']
                     },
                     index=['k0', 'k1', 'k2'])
right  = pd.DataFrame({
                        'C':['C0', 'C2', 'C3'],
                        'D':['D0', 'D2', 'D3']
                       },
                       index=['k0', 'k2', 'k3'])
print(left)
print(right)
```

```python
res = pd.merge(left, right, left_index=True, right_index=True, how='outer')
res
```

```python
boys = pd.DataFrame({'k':['k0', 'k0', 'k2'], 'age':[1,2,3]})
girls = pd.DataFrame({'k':['k0', 'k0', 'k3'], 'age':[4,5,6]})
print(boys)
print(girls)
```

```python
res = pd.merge(boys, girls, on='k', suffixes=['_boy', '_girl'], how='outer')
res 
```
