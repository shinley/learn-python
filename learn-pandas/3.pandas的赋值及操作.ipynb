```python
import pandas as pd
import numpy as np
```

```python
dates = np.arange(20170101, 20170107)
df1 = pd.DataFrame(np.arange(24).reshape(6,4), index=dates, columns=['A', 'B', 'C','D'])
df1
```

```python
df1.iloc[2,2]
```

```python
df1.iloc[2,2] = 100
df1
```

```python
df1.loc[20170102, 'B'] = 200
df1
```

```python
df1[df1.A>10] = 0
df1
```

```python
df1.A[df1.A==0] = 1
df1
```

```python
df1['E'] = 10
df1
```

```python
# 添加一列
df1['F'] = pd.Series([1,2,3,4,5,6], index=dates)
df1
```

```python
# 添加一行
df1.loc[20170107, ['A','B','C']] = [1,2,3]
df1
```

```python
s1 = pd.Series([1,2,3,4,5,6], index=['A','B','C','D','E','F'])
s1.name = 'S1'
df2 = df1.append(s1)
df2
```

```python
# 在第一列插入索引为G的df2中的E列
df1.insert(1,'G', df2['E'])
df1
```

```python
# 移动位置,先弹出列
g = df1.pop('G')
df1.insert(6,'G', g)
df1
```

```python
# 删 除列
del df1['G']
df1
```

```python
# 删除列
df2 = df1.drop(['A', 'B'], axis=1)
df2
```

```python
# 删除行
df3 = df1.drop([20170101, 20170102], axis=0)
df3
```
