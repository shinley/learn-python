```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
# 两行两列，在第一个位置 
plt.subplot(2,2,1)
        # x0, x1, y0, y1
plt.plot([0,1], [0, 1])
# 两行两列，在第二个位置 
plt.subplot(2,2,2)
plt.plot([0,1],[0,1])
# 两行两列，在第三个位置 
plt.subplot(2,2,3)
plt.plot([0,1],[0,1])
# 两行两列，在第四个位置 
plt.subplot(2,2,4)
plt.plot([0,1],[0,1])
plt.show()
```
