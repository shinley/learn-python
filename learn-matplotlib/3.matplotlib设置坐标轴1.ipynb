```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np. linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

# xy范围 
plt.xlim((-1, 2))
plt.ylim((-2, 3))
# xy描述
plt.xlabel('I am X')
plt.ylabel('I am Y')
plt.plot(x, y1, color='red', linewidth=1.0,  linestyle='--')
plt.plot(x, y2, color='blue', linewidth=2.0, linestyle='-')
plt.show()
```
